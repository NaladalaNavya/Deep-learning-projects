```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

The MNIST dataset (Modified National Institute of Standards and Technology dataset) is a large database of handwritten digits commonly used for training and testing in the field of machine learning. It contains 70,000 grayscale images of handwritten digits (0-9), each of size 28x28 pixels.

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
X_flatten_train= X_train.reshape(len(X_train),28*28)
X_flatten_test = X_test.reshape(len(X_test),28*28)
```

simple neural network

Define the Model:

Sequential API: For simple models, where layers are added one after another.

Functional API: For more complex models, such as those with multiple inputs/outputs or shared layers.

Model Subclassing: Allows for creating fully customized models by subclassing the Model class.

no hidden layer

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten

model = Sequential([
    Dense(10, input_shape=(784,), activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_flatten_train, y_train, epochs=5)
```

```python
model.evaluate(X_flatten_test,y_test)
```

```python
y_predicted = model.predict(X_flatten_test)
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

```python
y_predictes_label = [np.argmax(i) for i in y_predicted]
y_predictes_label[0:5]
```

```python
cm  = tf.math.confusion_matrix(y_test,y_predictes_label)
cm
```

```python
import seaborn as se
plt.figure(figsize=(10,7))
se.heatmap(cm,annot = True ,fmt = "d")
plt.xlabel("predicted")
plt.ylabel("Truth")
```

hidden layer

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten

model = Sequential([
    Dense(100, input_shape=(784,), activation='relu'),
    Dense(10, activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_flatten_train, y_train, epochs=5)
```

```python
model.evaluate(X_flatten_test,y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.optimizers  import SGD

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(100, input_shape=(784,), activation='relu'),
    Dense(10, activation='sigmoid')
])
model.compile(
    optimizer='SGD',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train, y_train, epochs=5)
```
